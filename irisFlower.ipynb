# Unzipping the file and getting csv

```python
import zipfile
import pandas as pd
with zipfile.ZipFile('irisReal.zip') as source:
  with source.open('iris.csv') as dataframe:
    df = pd.read_csv(dataframe)
    backup = df
print('done')
```
